** Hard: How many people have a credit card that expires in 2025? **

```python
cc_exp_2025 = list(df['CC Exp Date'])
# print(cc_exp_2025)
count = 0
for i in cc_exp_2025:
  if i.split("/")[1] == '25':
    # print(i)
    count += 1
count
```
