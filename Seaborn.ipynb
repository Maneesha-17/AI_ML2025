```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.info()
```

```python
tips.describe().T
```

# Bar plot
gives you aggregate data of a categorical feature

```python
sns.barplot(x ='sex', y = 'total_bill', data =tips, estimator = np.std)
```

#Countplot
gives you the number of occurrences,

```python
sns.countplot(x = 'sex', data = tips)
```

```python
tips.head()
```

# Boxplot & Violinplot
They are used to show the distribution of categorical data. a box plot (or box and whisker plot) shows the dist of quantitative data comparing between variables or across levels of categorical variable. The box shows the quartiles of the data while the whiskers extend to show the rest of the dist, except for points that are outliers.

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue = 'time',palette='rainbow')
```

# Violinplot
KDE - Kernel density estimation of the distribution.

```python
sns.violinplot(x='day', y = 'total_bill', hue = 'sex',data = tips, palette = 'coolwarm')
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, hue= 'sex', jitter = True, palette = 'rainbow')
```

```python
sns.violinplot(x="tip", y="day", data=tips,palette='rainbow')
sns.swarmplot(x="tip", y="day", data=tips,color='black',size=3)
```

# Matrix

Matrix plots allow you to plot data as color-encoded matrices and can also be used to indicate clusters within the data

```python
df = sns.load_dataset('flights')
df
```

```python
data = df.corr(numeric_only=True)
sns.heatmap(data)
```

```python
tips
```

```python
sns.heatmap(tips.corr(numeric_only=True), cmap = 'coolwarm', annot = True)
```

```python
pv=df.pivot_table(values = 'passengers', index = 'month', columns = 'year')
pv
```

```python
sns.heatmap(pv, linecolor = 'white', linewidths = 1)
```

```python
df.info()
```

```python
cluster1 = df.select_dtypes(include=np.int64).columns
```

```python
sns.clustermap(df[cluster1], cmap="coolwarm", standard_scale=1)
```

```python
iris = sns.load_dataset('iris')
iris.species.value_counts()
```

```python
iris
```

```python
sns.PairGrid(iris)
```

```python
s = sns.PairGrid(iris)
s.map(plt.scatter)
```

```python
s = sns.PairGrid(iris)
s.map_diag(plt.hist)
s.map_upper(plt.scatter)
s.map_lower(sns.barplot)
```

```python
sns.pairplot(iris, hue= 'species', palette = 'rainbow')
```

# FacetGrid
Creates grids of subplots, based on the values of categorical data.

```python
g = sns.FacetGrid(iris,col = 'species', height = 4, aspect = 1.2)
g.map(sns.scatterplot,'sepal_length', 'sepal_width')
```

```python
tips
```

```python
f = sns.FacetGrid(tips, col='smoker', row = 'day')
f.map(plt.hist,'total_bill')
```

# Regression plots
1. lmplot - linear reg plot

```python
sns.regplot(x = 'sepal_length', y = 'sepal_width', data = iris)
```
