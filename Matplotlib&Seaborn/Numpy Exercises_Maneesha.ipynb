#### Create a 4x4 array of random numbers between 0-10. Then subtract the mean of each row from the corresponding row using broadcasting.

```python
a = np.random.randint(0, 11, size = (4, 4))
a
```

```python
mean = np.mean(a, axis = 1)
mean
```

```python
after_subt = mean - a
after_subt
```
