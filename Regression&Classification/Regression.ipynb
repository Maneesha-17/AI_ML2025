```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
noise_scale = 100
number_of_samples = 50
x = 30*(np.random.rand(number_of_samples,1) - 0.5)
y = 5*x + 15*x**2 + 2*x**3 + noise_scale*np.random.randn(number_of_samples, 1)
plt.plot(x,y,'ro')
```

```python
x1 = x.ravel()
y1 = y.ravel()

z = np.polyfit(x1, y1, 1)

k = np.poly1d(z)
mse = np.square(np.subtract(y1,k(x1))).mean()
rmse = np.sqrt(mse)

print(f"Root Mean Squared Error: {rmse}")

r2 = r2_score(y1, k(x1))
print(f"R-squared (R²) Score for Linear Regression: {r2:.4f}")

curve_length = np.linspace(min(x1)-1,max(x1)+1)

plt.plot(x1, y1, 'ro', label = 'Data')
plt.plot(curve_length, np.polyval(z, curve_length),'g', label = 'Linear Regression')

plt.legend()
plt.show()
```
